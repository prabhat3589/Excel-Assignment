Question 1: What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly
explain with examples.

--> "Paste" applies a direct copy including all original data, formatting, formulas, and comments. "Paste Special," however, provides a dialog box or menu with options to selectively choose what to paste, such as only values, formulas, formats, or even link the pasted data to the original source.

**Paste**

Functionality: Copies an exact replica of the source, including all associated attributes.

Example: If you copy a cell with the formula =RANDBETWEEN(1,100), pasting it normally will result in a new cell containing the same formula, and any time the cell is changed, the formula will recalculate.

**Paste Special**

Functionality: Allows you to paste specific components of the copied content, offering fine-grained control over the destination cell.

Examples:

Paste Values: Copy a cell with a formula like =RANDBETWEEN(1,100) and use Paste Special to paste Values. The destination cell will show only the calculated number, not the formula, so it won't change.

Paste Formats: Copy a cell with a bold font and a red color, then use Paste Special to paste Formats. The destination cell will inherit the bold and red formatting but will not include any text or numbers.

Paste Link: Copy data and use the Paste Link option to create a dynamic link to the original data. Changes to the original data source will automatically update in the linked cell.

Transpose: If you have a horizontal list of data (in a row), you can use Paste Special > Transpose to paste it vertically (into a column), or vice versa.


Question 2: Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes'
in Excel.

--> **Freeze Panes**

This function locks specific rows and/or columns in place, keeping them visible as you scroll through the rest of the worksheet.

***Function: ***
Once you select a cell and freeze the panes, all rows above that cell and all columns to the left of it become fixed.

Freeze Top Row: Keeps the top row (Row 1) visible as you scroll down.

Freeze First Column: Keeps the first column (Column A) visible as you scroll horizontally.

Freeze Panes (Custom): Locks all rows above and all columns to the left of the active cell.

***Usefulness:***

Maintaining Context: Keeps important headers or labels visible at all times, preventing confusion when analyzing data far from the top or left.

Improved Navigation: Simplifies browsing large spreadsheets by eliminating the need to scroll back and forth to see row and column identifiers.

Data Analysis: Allows for easy comparison of data points with their corresponding labels, enhancing data analysis.

**Split Panes**

This function divides the Excel window into separate panes, each with its own scroll bar. This allows you to view different sections of the same worksheet simultaneously.

**Function: **
You can split the screen horizontally, vertically, or both to create up to four separate, independently scrollable panes.

Horizontal Split: Select a cell in column A and apply the Split command to create two horizontal panes.

Vertical Split: Select a cell in row 1 and apply the Split command to create two vertical panes.
Four-Way Split: Select any cell that is not in row 1 or column A and click Split to create four panes.

**Usefulness:**

Comparative Analysis: Compare data from different sections of the same worksheet side-by-side, which is highly useful for comparing sales figures from different years or regions.

Multitasking: Work on multiple parts of a spreadsheet at once without losing your place. For instance, you can enter data in one pane while viewing totals or summaries in another.

Flexible Viewing: Panes can be resized by dragging the split bar, allowing you to focus on the sections that are most important.

Question 3: Explain the difference between inserting a new row and inserting a new
column in Excel. Can you insert multiple rows or columns at once?

--> Inserting a row adds a horizontal set of blank cells, shifting existing data down, while inserting a column adds a vertical set of blank cells, shifting existing data to the right. You can insert multiple rows or columns at once by first selecting the number of existing rows or columns you wish to duplicate and then using the 'Insert' option from the right-click menu or the Home tab.

Question 4: What are logical functions in Excel? Provide examples of at least two
logical functions and their applications.

--> Logical functions in Excel perform tests on values and conditions, returning a specific result, often TRUE or FALSE, to help make decisions or perform actions based on the test outcome. Common logical functions include IF, which executes different actions based on a condition, and AND, which returns TRUE only if all specified conditions are true. For example, an IF function can determine if a student passed or failed a test, while an AND function can verify if an employee meets both a sales and attendance target to qualify for a bonus.

Examples of Logical Functions and Applications

**IF Function:**

Description: The IF function is a core decision-making function that checks a condition and returns one result if the condition is TRUE and another result if the condition is FALSE.

Formula: =IF(logical_test, value_if_true, value_if_false)

Application: Determine if a student passed or failed a course based on their score. If a student's score (in cell B2) is greater than 50, the formula =IF(B2 > 50, "Pass", "Fail") will display "Pass"; otherwise, it will display "Fail".

**AND Function:**

Description: The AND function checks multiple conditions simultaneously and returns TRUE only if all the specified conditions are true.

Formula: =AND(condition1, condition2, ...)

Application: Award a bonus only if an employee meets both sales and attendance criteria. The formula =AND(C2 >= 100, D2 >= 90) could be used in conjunction with an IF function; if the value in cell C2 (sales) is 100 or more AND the value in cell D2 (attendance) is 90 or more, then the employee qualifies for the bonus.

Question 5: Discuss the purpose of 'XLOOKUP' and how it differs from the traditional
'VLOOKUP' function.

--> Purpose of XLOOKUP

The XLOOKUP function in Excel is designed to search for a value in a range (lookup array) and return a corresponding value from another range (return array).

It was introduced to replace older lookup functions like VLOOKUP, HLOOKUP, and even LOOKUP, providing more flexibility and fewer limitations.

Key Differences Between XLOOKUP and VLOOKUP

VLOOKUP can only return a single value at a time, limiting its efficiency. On the contrary, XLOOKUP can return values from multiple columns in one go. This means you can use a single XLOOKUP formula instead of multiple VLOOKUPs.

Question 10: You're working with a dataset that contains missing values. As a Data
Scientist, explain how you'd detect and handle missing data using Excel.
Mention tools like:
● Go To Special
● ISBLANK
● COUNTBLANK

--> 1. Detecting Missing Values

In Excel, missing data usually appears as blank cells. You can detect them using different tools:

a) Go To Special

Select the dataset.

Press Ctrl + G (or use Home → Find & Select → Go To Special).

Choose Blanks → OK.

Excel will highlight all blank cells.

 This is useful if you want to quickly see where data is missing and take action (like filling, deleting, or replacing).

b) ISBLANK Function

The ISBLANK(cell) function checks if a specific cell is blank:

=ISBLANK(B2)


Returns TRUE if cell B2 is empty.

Returns FALSE if it contains data.

 You can apply this down a column to detect which rows have missing values.

c) COUNTBLANK Function

The COUNTBLANK(range) function counts the number of blank cells in a range:

=COUNTBLANK(B2:B20)


Returns the total number of empty cells in that range.

Helpful for quickly checking how much missing data you have.

2. Handling Missing Values

Once missing values are detected, you can decide how to handle them depending on the dataset and analysis needs:

Delete rows/columns with missing data → if only a few values are missing and they don’t matter.

Fill missing values manually → e.g., if you know the correct data.

Use statistical replacement → like replacing with the average, median, or a default value. For example:

Use =IF(ISBLANK(B2),AVERAGE($B$2:$B$20),B2) → replaces blanks with the average of the column.

Mark them clearly → replace blanks with "N/A" or "Missing" for reporting.

3. Summary

Detect:

Go To Special → highlight blanks.

ISBLANK → check cell by cell.

COUNTBLANK → count total missing values.

Handle:

Delete, replace, or impute missing values depending on context.
